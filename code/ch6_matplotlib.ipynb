## matplotlib 활용

### 기본적인 그리기 방법

```python
import numpy as np
from numpy import sin, cos
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#-pi부터 pi 까지 0.1을 한 구간으로 배열 생성
x = np.arange(-np.pi, np.pi, 0.1)
```

```python
#배열 x에 대해 sin(x)를 계산
y1 = sin(x)
```

```python
#배열 x에 대해 cos(x)를 계산
y2 = cos(x)
```

각각의 플로팅 작성, 레이블 인자 지정

```python
#x,y를 그리기
plt.plot(x, y1, label='sin')
plt.plot(x, y2, 'r*', markersize=10, label='cos')
# 범례를 그림
plt.legend(loc='best')

# 범례 두번째 방법
#plt.plot(x, y1)
#plt.plot(x, y2, 'r*', markersize=10)
#plt.legend(('sin','cos'),loc='best')

#각 축의 레이블을 그림
plt.xlabel('X-axis'); plt.ylabel('Y-axis')

#그리기 실행
plt.show()
```

## 서브 플롯

### subplot() 함수

```python
import matplotlib.pyplot as plt
#서브 플롯 생성
plt.subplot(2, 3, 1), plt.xticks([]), plt.yticks([])
#서브 플롯에 text 삽입
plt.text(0.5, 0.5, 'subplot(2,3,1)', ha='center', va='center', size=15)
plt.subplot(2, 3, 2), plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,3,2)', ha='center', va='center', size=15)
plt.subplot(2, 3, 3), plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,3,3)', ha='center', va='center', size=15)
plt.subplot(2, 3, 4), plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,3,4)', ha='center', va='center', size=15)
plt.subplot(2, 3, 5), plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,3,5)', ha='center', va='center', size=15)
plt.subplot(2, 3, 6), plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, 'subplot(2,3,6)', ha='center', va='center', size=15)
plt.tight_layout()
plt.show()
```

### GridSpec() 함수

```python
import matplotlib.gridspec as gs

G = gs.GridSpec(3, 3)
axes_1 = plt.subplot(G[0, :])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, '속틀1', ha='center', va='center', size = 22)
axes_2 = plt.subplot(G[1, :-1])
x = np.arange(-np.pi, np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.text(-1, 0, '속틀2', ha = 'center', va = 'center', size = 22)
axes_3 = plt.subplot(G[1:, -1])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, '속틀3', ha='center', va='center', size = 22)
axes_4 = plt.subplot(G[-1, 0])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, '속틀4', ha='center', va='center', size = 22)
axes_5 = plt.subplot(G[-1, -2])
plt.xticks([]), plt.yticks([])
plt.text(0.5, 0.5, '속틀5', ha='center', va='center', size = 22)

#plt.subplots_adjust(wspace = 0, hspace = 0) # 서브 플롯 간 간격

plt.show()
```

## 등고선 차트

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
#2차원 메시를 작성
y,x = np.mgrid[-2:2:.01, -3:3:.01]
pos = np.dstack((x,y))

#2변량 정규 분포로 2차원 분포 데이터를 생성
z = 15*(multivariate_normal.pdf(pos,[0,0],[[1,0],[0,1]])-
        multivariate_normal.pdf(pos, [1,1],[[1.5,0],[0,0.5]]))
```

```python
z.shape
```

```python
plt.contourf(x,y,z)
```

multivariate_normal.pdf(x, mean=None, cov=1, allow_singular=False)

```python
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', #origin (0,0)을 왼쪽 위, 아래중 어디에 놓을 것인지
                extent=(-3, 3, -2, 2)) #축의 범위
```

```python
#z의 값을 농담값으로 삼아 이미지를 나타냄
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', #origin (0,0)을 왼쪽 위, 아래중 어디에 놓을 것인지
                extent=(-3, 3, -2, 2)) #축의 범위

#등고선 표시
levels = np.arange(-2.5, 2.5, 0.5)
ctr = plt.contour(z,levels,  colors = 'k', origin = 'lower',
                  linewidths=2, extent=(-3, 3, -2, 2))
```

```python
#z의 값을 농담값으로 삼아 이미지를 나타냄
im = plt.imshow(z, interpolation = 'bilinear', origin = 'lower', #origin (0,0)을 왼쪽 위, 아래중 어디에 놓을 것인지
                extent=(-3, 3, -2, 2)) #축의 범위

#등고선 표시
levels = np.arange(-3, 3, 0.5)
ctr = plt.contour(z,levels,  colors = 'k', origin = 'lower',
                  linewidths=2, extent=(-3, 3, -2, 2))

#등고선에 레이블 인라인 표시
plt.gca().clabel(ctr, levels, inline=1, colors='black',fmt='%1.1f', fontsize=14)
plt.show()
```

## 3차원 플롯

```python
## 3차원 플롯
#데이터를 3차원 플롯으로 플로팅
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(1)    

#add_subplot()으로 3차원 플롯을 그려넣을 틀을 선정한다. 이 때,  
#    projection = 3d 옵션을 주어 3차원 플롯을 그리겠다고 선언한다
ax =fig.add_subplot(111, projection ='3d') 

#실제 데이터를 플로팅한다. cmap 옵션으로 초록색 스케일을 설정하였다
surf = ax.plot_surface(x,y,z, cmap=cm.Greens)
ax.set_zlim3d(-3.01, 3.01)
plt.colorbar(surf, shrink=0.5, aspect=10) #shrink: 그림에 대한 비율, aspect: 가로세로 비율
plt.show()
```

### 산점도

```python
import matplotlib.pyplot as plt
import numpy as np
# Reproducible random state
np.random.seed(19680801)    

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
# s는 마커의 크기, c는 마커의 색상, alpha는 투명도
plt.scatter(x, y, s=area, c=colors, alpha=.5)
plt.show()
```

### 파이차트

```python
# 기본 사용
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# 시작 각도와 방향 설정하기
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', 
        startangle=260, counterclock=False)
plt.show()
```

```python
# 중심에서 벗어나는 정도 설정
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False, explode=explode)
plt.show()
```

```python
# 색상지정
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, 
        counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```
